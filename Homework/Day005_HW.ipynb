2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests,json
popUrl='https://www.dcard.tw/_api/forums/pet/posts?popular=true'
npopUrl='https://www.dcard.tw/_api/forums/pet/posts?popular=false'
r=requests.get(popUrl)
response=json.loads(r.text)
print('ä¸€æ¬¡å›å‚³',len(response),'ç­†è³‡æ–™')
print('æ¯ç­†è³‡æ–™åŒ…å«ä»¥ä¸‹æ¬„ä½')
for a in response[0]:
    print(a)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
i=1
for a in response:
    print(str(response.index(a)+1)+'.')
    print('æ¨™é¡Œ:'+str(a['title']))
    print('è²¼æ–‡æ™‚é–“:'+str(a['createdAt']))
    print('ç•™è¨€äººæ•¸:'+str(a['commentCount']))
    print('æŒ‰è®šäººæ•¸:'+str(a['likeCount']))
    print('----------')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
url={'pop':popUrl,'npop':npopUrl}
count={'pop':{'CC':0,'LC':0},'npop':{'CC':0,'LC':0}}
for sort in ['pop','npop']:
    r=requests.get(url[sort])
    response=json.loads(r.text)
    tempCC=0
    tempLC=0
    for a in response:
        tempCC+=a['commentCount']
        tempLC+=a['likeCount']
    count[sort]['CC']=tempCC/len(response)
    count[sort]['LC']=tempLC/len(response)
        
print('ç†±é–€:\nå¹³å‡ç•™è¨€äººæ•¸:'+str(count['pop']['CC'])+'\nå¹³å‡æŒ‰è®šäººæ•¸:'+str(count['pop']['LC']))
print('-----------------')
print('éç†±é–€:\nå¹³å‡ç•™è¨€äººæ•¸:'+str(count['npop']['CC'])+'\nå¹³å‡æŒ‰è®šäººæ•¸:'+str(count['npop']['LC']))
```
